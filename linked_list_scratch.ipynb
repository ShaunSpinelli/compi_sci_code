```python
class Node:
    def __init__(self, value):
      self.value = value
      self.next = None
```

```python
class LinkedList:
  def __init__(self) -> None:
     self.head = None

  def add_node(self, data):
    node = Node(data)
    if self.head is None:
      self.head = node
      return
    last_node = self.head

    # go through nodes untill we find the end (.next is None)
    while last_node.next:
      last_node = last_node.next

    # update last node with new last node
    last_node.next = node

    return

  def remove_node(self, data):
    prev = None
    node =  self.head

    while node.next:
      if node.value == data:
        if prev is None: # node is head
          self.head = node.next
          return
        prev.next = node.next # node is in the middle
        return
      prev = node
      node = prev.next

    # node is last
    prev.next = None

  def print(self):
    node =  self.head

    ll = ""
    while node.next:
      ll += f"{node.value} -> "
      node = node.next

    ll += f"{node.value} -> null"
    print(ll)
```

```python
llist = LinkedList()
```

```python
llist.add_node(1)
```

```python
llist.add_node(2)
llist.add_node(3)
llist.add_node(4)
llist.add_node(5)
```

```python
llist.print()
```
